```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
print(data.shape)
data.head()
```

```python
m = 200000
y = np.zeros((m, 9))
i = 0
for Class_num in data['target']:
    y[i][int(Class_num.replace("Class_", "")) - 1] = 1
    i = i + 1
print(y[0:5,:])
```

```python
inp_df = data.drop(data.columns[[0,76]], axis=1)
inp_df.head()
```

```python
x = np.array(inp_df.values, 'float')
x[0:5, 0:10]
```

```python
# x = np.insert(x, 0, 1.0, axis=1)
# x[0:5, 0:10]
```

```python
def softmax(z):
    exp = np.exp(z - np.max(z))
    for i in range(len(z)):
        exp[i] /= np.sum(exp[i])
    return exp
```

```python
def pro_prob(w, b):
    pro_prob = x@w + b
    for i in range(m):
        for j in range(9):
            pro_prob[i][j] = max(min(pro_prob[i][j], 1 - 1e-15), 1e-15)
    return pro_prob
```

```python
def grad(w, b, alpha, iteration):
    
    
    
    J_history = np.zeros([iteration,])
    
    for it in range(iteration):
        z = pro_prob(w, b)
        y_hat = softmax(z)
        
        w_grad = (1/m)*np.dot(x.T, (y_hat - y))
        b_grad = (1/m)*np.sum(y_hat - y)
        
        w = w - alpha * w_grad
        b = b - alpha * b_grad
        
        loss = -np.sum(y*np.log(z))/m
        
        J_history[it] = loss
        
    return w, b, J_history
```

```python
w = np.zeros([75, 9])
b = np.zeros([1, 9])
w, b, l = grad(w, b, 0.01, 30)
```

```python
print(l)
```

# Training Done
Now we have required **w** and **b** for optimal solution. Lets import test cases and predict the output.

```python
test_data = pd.read_csv('test.csv')
test_data.head()
```

```python
def predict_prob(X, w, b):
    pro_prob = X@w + b
    for i in range(X.shape[0]):
        for j in range(9):
            pro_prob[i][j] = max(min(pro_prob[i][j], 1 - 1e-15), 1e-15)
    return pro_prob
```

```python
input_data = test_data.drop(test_data.columns[[0]], axis=1)
```

```python
input_data.head()
```

```python
X = np.array(input_data.values, 'float')
X[0:5, 0:10]
```

```python
pre_out = predict_prob(X, w, b)
```
